### 25.	Get Designation, total salary spend by the company with respect to each E.g.
Designation Name		Total Salary
<br> Mobile Software Engineer	622100.00


```python
Select_Desig_Salary_Spend = pd.merge(Employee_Details, Country, left_on = 'emp_country_id', right_on = 'country_id', how='left')
Select_Desig_Salary_Spend = pd.merge(Select_Desig_Salary_Spend, Designation, left_on = 'desig_id', right_on = 'desig_id', how='left')
Select_Desig_Salary_Spend = pd.merge(Select_Desig_Salary_Spend, Salary, left_on = 'emp_id', right_on = 'emp_id', how='right')
Select_Desig_Salary_Spend = Select_Desig_Salary_Spend[['emp_id', 'desig_description', 'emp_salary_change_year', 'emp_salary']]
Select_Desig_Salary_Spend['date_shift'] = Select_Desig_Salary_Spend.groupby(['emp_id'])['emp_salary_change_year'].apply(lambda x: x.shift(-1))
Select_Desig_Salary_Spend.ix[Select_Desig_Salary_Spend['date_shift'].isnull(), 'date_shift'] = date.today()
Select_Desig_Salary_Spend['date_diff'] = (Select_Desig_Salary_Spend['date_shift'] - Select_Desig_Salary_Spend['emp_salary_change_year'] + timedelta(days=1))/np.timedelta64(1, 'Y')
Select_Desig_Salary_Spend['Total Salary Spend'] = Select_Desig_Salary_Spend['date_diff'] * Select_Desig_Salary_Spend['emp_salary']
Select_Desig_Salary_Spend = Select_Desig_Salary_Spend.sort_values(by='desig_description')
Select_Desig_Salary_Spend = Select_Desig_Salary_Spend.groupby(['desig_description']).agg({'Total Salary Spend': 'sum'})
Select_Desig_Salary_Spend
```

### 30.	Select first name, sum of salary from employee details and salary table for all employees even if they didn't get salary and set salary amount as 0 for those employees who didn't get salary. 
Hint: use left outer join


```python
Select_Emp_Salary_Sum = pd.merge(Employee_Details, Country, left_on = 'emp_country_id', right_on = 'country_id', how='left')
Select_Emp_Salary_Sum = pd.merge(Select_Emp_Salary_Sum, Designation, left_on = 'desig_id', right_on = 'desig_id', how='left')
Select_Emp_Salary_Sum = pd.merge(Select_Emp_Salary_Sum, Salary, left_on = 'emp_id', right_on = 'emp_id', how='left')
Select_Emp_Salary_Sum = Select_Emp_Salary_Sum[['emp_first_name', 'desig_description', 'emp_salary_change_year', 'emp_salary']]
Select_Emp_Salary_Sum['date_shift'] = Select_Emp_Salary_Sum.groupby(['emp_first_name'])['emp_salary_change_year'].apply(lambda x: x.shift(-1))
Select_Emp_Salary_Sum.ix[Select_Emp_Salary_Sum['date_shift'].isnull(), 'date_shift'] = date.today()
Select_Emp_Salary_Sum['date_diff'] = (Select_Emp_Salary_Sum['date_shift'] - Select_Emp_Salary_Sum['emp_salary_change_year'] + timedelta(days=1))/np.timedelta64(1, 'Y')
Select_Emp_Salary_Sum['Total Salary Gain'] = Select_Emp_Salary_Sum['date_diff'] * Select_Emp_Salary_Sum['emp_salary']
Select_Emp_Salary_Sum = Select_Emp_Salary_Sum.groupby(['emp_first_name']).agg({'Total Salary Gain':'sum'})
Select_Emp_Salary_Sum
```

### 35.	Display employee name, and average salary with the help of a user defined function.

```python
def Average_Emp_Salary():

    Select_Emp_Salary_Avg = pd.merge(Employee_Details, Country, left_on = 'emp_country_id', right_on = 'country_id', how='left')
    Select_Emp_Salary_Avg = pd.merge(Select_Emp_Salary_Avg, Designation, left_on = 'desig_id', right_on = 'desig_id', how='left')
    Select_Emp_Salary_Avg = pd.merge(Select_Emp_Salary_Avg, Salary, left_on = 'emp_id', right_on = 'emp_id', how='right')
    Select_Emp_Salary_Avg = Select_Emp_Salary_Avg[['emp_first_name', 'emp_last_name', 'emp_middle_name', 'country_name', 'desig_name', 'emp_salary']].sort_values(by='desig_name')
    Select_Emp_Salary_Avg = Select_Emp_Salary_Avg.groupby(['emp_first_name', 'emp_last_name', 'emp_middle_name', 'country_name', 'desig_name']).agg({'emp_salary':'mean'}).reset_index()

    return(Select_Emp_Salary_Avg)

Select_Emp_Salary_Avg = Average_Emp_Salary()
Select_Emp_Salary_Avg
```
